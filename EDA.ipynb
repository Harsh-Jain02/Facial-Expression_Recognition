```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_folder = "./FER/train"
emotions = os.listdir(train_folder)
print(emotions)
```

```python
images = os.listdir(os.path.join(train_folder, emotions[0]))
```

```python
for i in enumerate(emotions):
    print(i)
```

```python
emot = {}
for emotion in emotions:
    emot[emotion] = len(os.listdir(os.path.join(train_folder, emotion)))
    print(emot[emotion])
```

```python
plt.plot(emot.keys(), emot.values())
plt.ylim(0,)
plt.title("Distribution in training")
plt.show()
```

```python
np.array(list(emot.values()))/emot["disgust"]
```

```python
test_folder = "./FER/test"
emotions = os.listdir(test_folder)
```

```python
emot_t = {}
for emotion in emotions:
    emot_t[emotion] = len(os.listdir(os.path.join(test_folder, emotion)))
    print(emot_t[emotion])

plt.plot(emot_t.keys(), emot_t.values())
plt.ylim(0,)
plt.title("Distribution in testing")
plt.show()
```

```python
plt.plot(emot_t.keys(), emot_t.values())
plt.plot(emot.keys(), emot.values())
plt.ylim(0,)
plt.title("Distribution in testing")
plt.show()
```

```python
plt.plot(emot_t.keys(), np.array(list(emot_t.values()))/emot_t["disgust"], label = "test")
plt.plot(emot.keys(), np.array(list(emot.values()))/emot["disgust"], label = "train")
plt.ylim(0,)
plt.legend()
plt.title("Distribution in testing")
plt.show()
```
